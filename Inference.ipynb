```python
#import tensorflow
import tensorflow
```

```python
#read the model trained on 20 images from github
import urllib.request

urllib.request.urlretrieve(
        'http://github.com/venkatkulkarni/CatsVsDogs/blob/main/dog_cat_model_Final_20Images_dataset.h5?raw=true', 'model.h5')
```

```python
#set the model path 
model_file_path = './model.h5'
model_checkpoint = tensorflow.keras.callbacks.ModelCheckpoint(
    model_file_path,
    monitor="val_loss",
    save_best_only=True,
    mode="min",
)
```

```python
#load the pre trained model which is trained on 20 images
from tensorflow.keras.models import load_model
classifier = load_model(model_file_path)
```

```python
#import PIL as it is needed to read the images from github for inference
from PIL import Image
from urllib import request
from io import BytesIO
```

```python
#Prediction of image
%matplotlib inline
import tensorflow
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
for i in range(5):
    url = "https://github.com/venkatkulkarni/CatsVsDogs/blob/main/Testdata/{}.jpg?raw=true".format(i+1)
    res = request.urlopen(url).read()
    Sample_Image = Image.open(BytesIO(res)).resize((64,64))
    img = image.img_to_array(Sample_Image)
    img = img/255
    # create a batch of size 1 [N,H,W,C]
    img = np.expand_dims(img, axis=0)
    prediction = classifier.predict(np.asarray(img), batch_size=None,steps=1) #gives all class prob.
    if(prediction[:,:]>0.5):
        value ='Dog :%1.2f'%(prediction[0,0])
        plt.text(20, 62,value,color='red',fontsize=18,bbox=dict(facecolor='white',alpha=0.8))
    else:
        value ='Cat :%1.2f'%(1.0-prediction[0,0])
        plt.text(20, 62,value,color='red',fontsize=18,bbox=dict(facecolor='white',alpha=0.8))

    plt.imshow(np.squeeze(img))
    plt.show()
```

```python
#test the model on sample 20 images and we can observe that model which is trained on 20 images is not prediciting accurately 
fig=plt.figure(figsize=(15, 6))
columns = 7
rows = 3
for i in range(20):
    fig.add_subplot(rows, columns, i+1)
    url = "https://github.com/venkatkulkarni/CatsVsDogs/blob/main/Testdata/{}.jpg?raw=true".format(i+1)
    res = request.urlopen(url).read()
    Sample_Image = Image.open(BytesIO(res)).resize((64,64))
    img = image.img_to_array(Sample_Image)
    img = img/255
    img = np.expand_dims(img, axis=0)
    prediction = classifier.predict(img, batch_size=None,steps=1) #gives all class prob.
    if(prediction[:,:]>0.5):
        value ='Dog :%1.2f'%(prediction[0,0])
        plt.text(20, 58,value,color='red',fontsize=10,bbox=dict(facecolor='white',alpha=0.8))
    else:
        value ='Cat :%1.2f'%(1.0-prediction[0,0])
        plt.text(20, 58,value,color='red',fontsize=10,bbox=dict(facecolor='white',alpha=0.8))
    plt.imshow(Sample_Image)
```
